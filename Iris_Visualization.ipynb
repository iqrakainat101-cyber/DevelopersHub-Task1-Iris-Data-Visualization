```python
# ==========================================
# TASK 1: IRIS DATASET EXPLORATION
# DevelopersHub AI/ML Internship
# ==========================================

# Import required libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

# Display plots in notebook
%matplotlib inline
```

```python
# Problem Statement
print("""
Task 1: Exploring and Visualizing the Iris Dataset

Objective:
Learn how to load, inspect, analyze, and visualize a dataset
to understand trends, distributions, and relationships among features.
""")
```

```python
# Load Dataset
# Load Iris dataset from sklearn

iris = load_iris()

# Convert to DataFrame

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

# Add species column

df['species'] = iris.target

# Convert numeric target to names

species_names = {
    0: 'Setosa',
    1: 'Versicolor',
    2: 'Virginica'
}

df['species'] = df['species'].map(species_names)

print("Dataset Loaded Successfully!")
```

```python
# Display Dataset Information
# Shape of dataset

print("Dataset Shape:")
print(df.shape)

print("\n")
print("="*50)

# Column names

print("Column Names:")
print(df.columns)

print("\n")
print("="*50)

# First five rows

print("First Five Rows:")
display(df.head())
```

```python
# Dataset Information
print("Dataset Information")
print("="*50)

df.info()
```

```python
# Summary Statisics
print("Summary Statistics")
print("="*50)

display(df.describe())
```

```python
# Check Missing Values
print("Missing Values")
print("="*50)

print(df.isnull().sum())
```

```python
# Scatter Plot
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='sepal length (cm)',
    y='petal length (cm)',
    hue='species',
    s=100
)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# Observation
print("""
Observation:
Setosa species is clearly separated from the others.
Petal length increases with sepal length.
""")
```

```python
# Pair Plot
sns.pairplot(
    df,
    hue='species'
)

plt.show()
```

```python
# Histograms
# Histograms for all numerical features

df.hist(
    figsize=(12,8),
    bins=15
)

plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
# Observation
print("""
Observation:
Petal measurements show clear separation among species.
Sepal features have more overlap.
""")
```

```python
# Box Plot for Outliners Detections
features = iris.feature_names

for feature in features:

    plt.figure(figsize=(6,4))

    sns.boxplot(
        x=df[feature]
    )

    plt.title(f"Box Plot of {feature}")
    plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))

correlation = df.drop('species', axis=1).corr()

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Observation
print("""
Observation:
Petal length and petal width have a strong positive correlation.
""")
```

```python
# Species Distribution
plt.figure(figsize=(6,4))

sns.countplot(
    x='species',
    data=df
)

plt.title("Species Distribution")
plt.show()
```

```python
# Final Insights
print("""
FINAL CONCLUSION

1. The Iris dataset contains 150 samples and 5 columns.

2. There are no missing values.

3. Setosa is easily distinguishable from other species.

4. Petal length and petal width are strongly correlated.

5. Histograms reveal different distributions for each feature.

6. Box plots show very few outliers.

7. Scatter plots clearly show relationships between flower measurements.

8. The dataset is clean and suitable for machine learning applications.
""")
```

```python
# Save Dataset
df.to_csv("iris_dataset.csv", index=False)

print("Dataset saved successfully.")
```
